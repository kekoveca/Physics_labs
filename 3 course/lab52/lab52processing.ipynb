```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style= 'whitegrid', font_scale= 2)
```

```python
calibr_x = np.array([1938, 2221, 2417, 3092])
calibr_y = np.array([4.784, 5.490, 6.002, 7.687])
```

```python
plf = np.polyfit(calibr_x, calibr_y, 1)

x = np.linspace(1000, 3500, 10)
y = np.polyval(plf, x)
```

```python
plt.figure(figsize = (16,10), dpi= 250)

plt.scatter(calibr_x, calibr_y, c= 'black', s = 50)
plt.plot(x, y, label = f'$E = {plf[0]:.5}N {plf[1]:.5f}$', c = 'red')
plt.ylabel('E, МэВ')
plt.xlabel('N, номер канала')
plt.legend(loc = 'best')
plt.savefig('plot_calib')
```
