```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', font_scale=1.5)
```

```python
## Калибровка
fermi = np.array([ 269.4911, 231.3438, 218.5779, 222.7088, 212.0592, 199.5748, 185.6918, 158.5528, 144.0207, 116.3015, 112.6497, 85.9853, 67.1818, 40.4816, 28.5734, 31.2308, 60.4808, 69.7305, 71.3822, 66.5784, 49.3847, 29])
T = np.array([34.8, 53.4, 75.4, 100.3, 127.9, 157.8, 189.6, 223.3, 258.4, 294.9, 332.5, 371.1, 410.6, 450.9, 491.8, 533.3, 575.4, 617.9, 622.2, 628.6, 639.4, 660.9])
```

```python
pn = np.polyfit(T[:-7], fermi[:-7], 1)
fermi_new = np.polyval(pn, T[:-5])
root = np.roots(pn)[0]
```

```python
plt.figure(figsize = (16,10), dpi= 250)

plt.scatter(T, fermi)
plt.plot(T[:-5], fermi_new, linestyle= '-')
plt.plot([root, root], [-20, 300], linestyle = '--', label = f'$E_{{max}} = {root:.0f} \\text{{ кэВ}}$')
plt.ylim(-20, 300)
plt.ylabel('$\\frac{\sqrt{N(p)}}{p^{3/2}}$')
plt.xlabel('$T, keV$')
plt.legend(loc = 'lower left')
plt.savefig('plot')
```
