```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style= 'whitegrid', font_scale= 2)
```

```python
calibr_x = np.array([749.0, 1774.2, 952.9])
calibr_y = np.array([511, 1275, 662])
```

```python
plf = np.polyfit(calibr_x, calibr_y, 1)

x = np.linspace(500, 2000, 10)
y = np.polyval(plf, x)
```

```python
plt.figure(figsize = (16,10), dpi= 250)

plt.scatter(calibr_x, calibr_y, c= 'black', s = 50)
plt.plot(x, y, label = f'$E = {plf[0]:.5}N {plf[1]:.5f}$', c = 'red')
plt.ylabel('E, ')
plt.xlabel('N,  ')
plt.legend(loc = 'best')
plt.savefig('plot_calib')
```
