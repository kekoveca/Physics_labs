```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_theme(style='darkgrid', font_scale=1.5)
```

```python
data = pd.read_excel('lab12data.xlsx')
```

```python
phis = np.array(data.iloc[0:10, 0])
ns = np.array(data.iloc[0:10, 1])
```

```python
x_data = 1 - np.cos(np.deg2rad(phis))
y_data = 1 / ns

plf = np.polyfit(x_data, y_data, 1, full= True)
Pn = plf[0]
y_approx = np.polyval(Pn, x_data)

N0 = Pn[1]
N90 = np.polyval(Pn, x_data[8])
```

```python
errs = 0.01 * ns
errs_new = errs / ((ns) ** 2) 
```

```python
plt.figure(figsize= (12,10))
plt.scatter(x_data, y_data, c= 'black', s = 15)
plt.errorbar(x_data, y_data, errs_new, fmt = 'none', c = 'black')
plt.plot(x_data, y_approx, c = 'red')
plt.scatter([0,1], [N0, N90], marker= '*', s = 550, c = 'green', label = f'1/N(0) = {N0:.5f} $\pm$ {2*errs_new[0]:.5f} \n1/N(90) = {N90:.5f} $\pm$ {2*errs_new[8]:.5f}')
plt.legend()
plt.xlabel('$1 - cos(\\theta) $')
plt.ylabel('$ 1 / N(\\theta) $')
plt.savefig('plot.png')
```

```python
n90 = 1 / N90
n0 = 1 / N0

E = 662 * (n90 / (n0 - n90))

print(n0, n90, E)
print(errs[0]*2, errs[8]*2, )
```
