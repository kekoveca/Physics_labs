```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_theme(style='darkgrid', font_scale=1.5)
```

```python
data = pd.read_excel('lab12data.xlsx')
```

```python
phis = np.array(data.iloc[0:12, 0])
ns = np.array(data.iloc[0:12, 1])
```

```python
x_data = 1 - np.cos(np.deg2rad(phis))
y_data = 1 / ns

plf = np.polyfit(x_data, y_data, 1, full= True)
Pn = plf[0]
y_approx = np.polyval(Pn, x_data)
```

```python
plt.figure(figsize= (12,10))
plt.scatter(x_data, y_data, c= 'black')
plt.plot(x_data, y_approx, c = 'red')
plt.legend()
plt.xlabel('$1 - cos(\\varphi), deg $')
plt.ylabel('$ 1 / N(\\varphi) $')
plt.savefig('plot.png')
```
