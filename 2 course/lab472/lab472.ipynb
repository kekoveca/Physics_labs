```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(palette='deep', style= 'darkgrid', font_scale=1.5)
```

```python
r = np.array([3, 4.2, 5, 5.8, 6.5, 7.3, 7.8, 8.3])**2
m = np.arange(1, len(r) + 1)

plf = np.polyfit(m, r, 1, full= True)
r_squared = np.polyval(plf[0], m)
err = float(plf[1]/len(m))
err
```

```python
plt.figure(figsize = (12,8))
plt.scatter(m, r, c = 'black', alpha= 0.9)
plt.plot(m, r_squared, label = f'$k = {plf[0][0]:.1f} \pm {err:.1f}$')
plt.legend()
plt.xlabel('$m$')
plt.ylabel('$r^2, ^2$')
plt.title('$r^2 = f(m)$')
plt.savefig('plot1.png')
```
