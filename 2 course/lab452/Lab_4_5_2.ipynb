```python
import math as m
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
import warnings
warnings.filterwarnings("ignore")

x1 = np.array([60, 50, 40, 30, 20,10, 0, 100, 45, 180, 170, 160, 150, 130])
y1 = np.array([0.26, 0.32, 0.4, 0.39, 0.37, 0.33, 0.28, 0.08, 0.38, 0.2, 0.16, 0.12, 0.09, 0.05])
#y1 = np.array([0.64,0.65, 0.67, 0.64, 0.57, 0.51, 0.47, 0.6, 0.51, 0.41, 0.43, 0.32])
x11 = (x1/180)*np.pi
#Погрешность аппроксимации
yerr11 = 0
yerr12 = 0.002
yerr122 = 0.004
dy1 = yerr11

def mapping(x, a, b, c, d): 
    return (a*(np.cos(d*x + b))**2 + c)
popt, _ = optimize.curve_fit(mapping, x11, y1)
a, b, c, d = popt  
print("Коэффициент: k1 =", a, " ", b, " ", c, " ", d)

fig, ax = plt.subplots(figsize=(8, 7), dpi=150)
plt.title("Зависимость видности $\gamma$ от угла поляризации $\u03B1$",color="black", fontweight="bold") # заголовок
plt.xlabel("Угол, $\u03C6$ [рад]") # ось абсцисс
plt.ylabel("Видность, $\gamma_3$ ") # ось ординат  

plt.minorticks_on()
plt.grid(which='major',
        color = 'grey', 
        linewidth = 0.7)
plt.grid(which='minor', 
        color = 'grey', 
        linestyle = ':')
x0 = np.linspace(0,m.pi)
plt.errorbar(x11,y1, yerr=dy1,fmt='.',ecolor='k',color='navy',elinewidth=0.5, capsize = 3, label = 'Экспериментальные измерения - спектр')
#ax.plot(x0, (0.35*(np.cos(0.019*x0-0.8))**2 + 0.05 ), "blue", label = 'МНК-спектр', linewidth = 1)
ax.plot(x0, (a*(np.cos(d*x0 + b))**2 + c), "blue", label = 'Аппроксимация $cos^2(x)$', linewidth = 1)
#ax.plot(x1, a2*x1, "darkorange", label = 'МНК-микроскоп', linewidth = 1)
#plt.errorbar(x1,y2, yerr=dy1,fmt='.',ecolor='k',color='red',elinewidth=0.5, capsize = 3, label = 'Экспериментальные измерения - микроскоп')
ax.legend()


plt.text(1.33, 0.35, "$y = {}*cos^2({}x+{}) + {}$".format(round(a, 1), round(d, 1), round(b, 1), round(c, 1)), bbox=dict(boxstyle="round,pad=0.3", fc="lightblue", ec="darkblue", lw=1))
#plt.text(0.3, 0.123, "$k = {}\u00B1{}$".format(round(a2, 3), round(yerr122, 4)), bbox=dict(boxstyle="round,pad=0.3", fc="orange", ec="darkred", lw=1))
#plt.text(500, 0.1, "$b = {}$".format(round(p1[1], 3)), bbox=dict(boxstyle="round,pad=0.3", fc="lightblue", ec="darkblue", lw=1))

plt.savefig("plot4--5--2-1")
```
