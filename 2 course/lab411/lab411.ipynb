```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_excel('lab411.xlsx')
df.head(20)
```

```python
a1 = np.array(df.iloc[2:8, 0])
b1 = np.array(df.iloc[2:8, 1])
f1 = np.array(df.iloc[2:8, 2])

yspace = 1/a1
xspace = 1/b1
finv = np.mean(1/f1)

poly = np.polyfit(list(xspace), list(yspace), 1)
xpacenew = np.linspace(np.min(xspace), np.max(xspace), 100)
yspacenew = np.polyval(poly, xpacenew)
print(poly)
```

```python
plt.figure(dpi = 400)
plt.scatter(xspace, yspace)
plt.plot(xpacenew, yspacenew, alpha = 0.7, label = f'f = {1/poly[1]:.2f} +- 0.6')
plt.ylabel('1/a')
plt.xlabel('1/b')
plt.legend()
plt.grid(alpha = 0.5)
plt.savefig('plot1.png')
```
