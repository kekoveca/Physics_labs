```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', font_scale=1.7)
```

```python
m = np.arange(1, 7)
xm = np.array([0.22, 0.47, 0.71, 0.94, 1.18, 1.36])

plf = np.polyfit(m, xm, 1)
y = np.polyval(plf, m)
```

```python
plt.figure(figsize=(12, 10))
plt.scatter(m, xm)
plt.plot(m, y, c = 'r')
plt.xlabel('m')
plt.title(' $X_m$(m)')
plt.ylabel('$X_m$')
plt.savefig('Fraunhofer.jpg')
```
