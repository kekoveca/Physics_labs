```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_theme(style='darkgrid', font_scale=1.5)
```

```python
data = pd.read_excel('lab436.xlsx')

grid_2 = np.array(data.iloc[16:21, 2], dtype= float)
grid_3 = np.array(data.iloc[17:20, 3], dtype= float)
grid_4 = np.array(data.iloc[16:18, 4], dtype= float)
z_2 = np.arange(1, len(grid_2) + 1)
z_3 = np.arange(1, len(grid_3) + 1)

plf2 = np.polyfit(z_2, grid_2, 1, full= True)
plf3 = np.polyfit(z_3, grid_3, 1, full= True)

y2 = np.polyval(plf2[0], z_2)
y3 = np.polyval(plf3[0], z_3)

er2 = plf2[1]/len(z_2)
er3 = plf3[1]/len(z_3)

k2 = plf2[0][0]
k3 = plf3[0][0]
```

```python
plt.figure(figsize= (12,10))
plt.scatter(z_2, grid_2, c = 'b', label = f'Решётка 2, k = {k2:.2f} ')
plt.errorbar(z_2, y2, c = 'black', capsize= 5)
plt.scatter(z_3, grid_3, c = 'r', label = f'Решётка 3, k = {k3:.2f} ')
plt.errorbar(z_3, y3, c = 'black', capsize= 5)
plt.legend(loc = 'upper left')
plt.xlabel('n')
plt.ylabel('$z_n$')
plt.savefig('plot1.png')
```

```python
lb = 510 * 10**(-6)

d2 = np.sqrt(lb*k2/2)
d3 = np.sqrt(lb*k3/2)
```
