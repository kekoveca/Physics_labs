```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from math import sqrt

x1 = np.array([0.57, 1.15, 1.72, 2.31, 2.89, 3.45])
y1 = np.array([2.3, 4.7, 7, 10, 12.7, 15])

x2 = np.array([0.59, 1.16, 1.73, 2.31, 2.88, 3.46])
y2 = np.array([3, 5.3, 8, 11, 13.3, 15.7])

#  
p1 = np.polyfit(x1,y1,1);
P01 = np.polyval(p1,x1);

# 
yerr11 = 0.0791 #
yerr12 = 0.05 #
dy1 = sqrt(yerr11**2 + yerr12**2)

#  
p2 = np.polyfit(x2,y2,1);
P02 = np.polyval(p2,x2);

# 
yerr21 = 0.0883 #
yerr22 = 0.05 #
dy2 = sqrt(yerr21**2 + yerr22**2)

fig, ax = plt.subplots(figsize=(8, 6), dpi=150)
plt.title("  $B_ = f(I)$",color="black", fontweight="bold") # 
plt.xlabel("    , I []") #  
plt.ylabel("    , B []") #    

plt.minorticks_on()
plt.grid(which='major',
        color = 'k', 
        linewidth = 0.7)
plt.grid(which='minor', 
        color = 'k', 
        linestyle = ':')

plt.errorbar(x1,y1, yerr=dy1,fmt='.',ecolor='r',color='green',elinewidth=1, capsize = 3, label = 'B(I),  ')
ax.plot(x1, P01, "g-", label = ',  ', linewidth = 1)
plt.errorbar(x2,y2, yerr=dy2,fmt='.',ecolor='r',color='blue',elinewidth=1, capsize = 3, label = 'B(I),  ')
ax.plot(x2, P02, "b-", label = ',  ', linewidth = 1)
ax.legend()

plt.text(2.6, 7, "$k = {}$".format(round(p1[0], 3)), bbox=dict(boxstyle="round,pad=0.3", fc="lightgreen", ec="darkgreen", lw=1))
plt.text(2.55, 5, "$b = {}$".format(round(p1[1], 3)), bbox=dict(boxstyle="round,pad=0.3", fc="lightgreen", ec="darkgreen", lw=1))
plt.text(2.6, 3, "$\\sigma_B$ = {}".format(round(dy1, 3)), bbox=dict(boxstyle="round,pad=0.3", fc="lightgreen", ec="darkgreen", lw=1))
plt.text(3.1, 7, "k = {}".format(round(p2[0], 3)), bbox=dict(boxstyle="round,pad=0.3", fc="lightblue", ec="darkblue", lw=1))
plt.text(3.1, 5, "b = {}".format(round(p2[1], 3)), bbox=dict(boxstyle="round,pad=0.3", fc="lightblue", ec="darkblue", lw=1))
plt.text(3.1, 3, "$\\sigma_B$ = {}".format(round(dy2, 3)), bbox=dict(boxstyle="round,pad=0.3", fc="lightblue", ec="darkblue", lw=1))

plt.savefig("plot3-3-1--1")
```
