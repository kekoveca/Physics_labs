```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', font_scale=1.7)

m = np.array([1.1, 1.2, 1.3, 1.4, 1.5, 1.6])**-1
xm = np.array([1495, 1385, 1281, 1190, 1114, 1030])

plf = np.polyfit(m, xm, 1)
y = np.polyval(plf, m)

plt.figure(figsize=(12, 10))
plt.scatter(m, xm)
plt.plot(m, y, c = 'r')
plt.xlabel('$1/ \\nu $')
plt.title(' $\Lambda(1/ \\nu )$')
plt.ylabel('$\Lambda$')
plt.savefig('plot1.jpg')
print(plf)
```
